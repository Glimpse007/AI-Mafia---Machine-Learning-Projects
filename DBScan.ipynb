```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_circles,make_moons
```

```python
k,t = make_moons(n_samples=200,noise=0.10)
```

```python
plt.scatter(k[:,0],k[:,1],c=t)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
```

```python
km.fit(k)
```

```python
centers = km.cluster_centers_
label = km.labels_

plt.scatter(k[:,0],k[:,1],c=label)
plt.scatter(centers[:,0],centers[:,1],color='red')
plt.show()
```

DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps=0.2001,min_samples=5)
```

```python
dbs.fit(k)
```

```python
ypred = dbs.fit_predict(k)
```

```python
plt.scatter(k[:,0],k[:,1],c=ypred)
plt.show()
```
